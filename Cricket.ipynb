import zipfile
import pandas as pd
import os

# Path to the zip file
zip_file_path = "DataSet.zip"

# Create a directory to extract files
extracted_files_dir = "extracted_files"
os.makedirs(extracted_files_dir, exist_ok=True)

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_files_dir)

# Initialize an empty list to store DataFrames
data_frames = []

# List all files in the extracted directory
extracted_files = os.listdir(extracted_files_dir)

# Read each xlsx file using pandas
for file_name in extracted_files:
    if file_name.endswith('.xlsx'):
        # Construct the full path to the xlsx file
        file_path = os.path.join(extracted_files_dir, file_name)
        
        # Read the xlsx file into a DataFrame
        df = pd.read_excel(file_path)
        
        # Add a new column with the file name
        df.insert(0, 'FileName', file_name)
        
        # Append the DataFrame to the list
        data_frames.append(df)

# Check if any DataFrames were successfully read
if data_frames:
    # Concatenate all DataFrames into a single DataFrame
    df = pd.concat(data_frames, ignore_index=True)
else:
    print("No valid Excel files found in the extracted directory.")

# Now you have a DataFrame containing data from all extracted files
# You can access the file names using df['FileName'].unique()
